### Dataset Penjualan

```python
#Mengimpor pustaka atau modul bernama pandas ke dalam program Python
import pandas as pd

#Membaca file Excel menggunakan pandas
data = pd.read_excel('Data Penjualan 2022-2023 - FINAL.xlsx')

data.info()
```

```python
#Menampilkan beberapa baris pertama dari dataset
data.head(5)
```

### Penjualan Tahun 2022 per Bulan

```python
#Menghitung jumlah penjualan per bulan
data['tanggal'] = pd.to_datetime(data['tanggal'])
total_penjualan_per_bulan_2022 = []

for bulan_yang_diinginkan in range(1, 13):
    penjualan_per_bulan = data[(data['tanggal'].dt.month == bulan_yang_diinginkan) & (data['tanggal'].dt.year == 2022)]
    total_penjualan_per_bulan = penjualan_per_bulan['jumlah penjualan'].sum()
    total_penjualan_per_bulan_2022.append({'Bulan': bulan_yang_diinginkan, 'Total Penjualan': total_penjualan_per_bulan})

#Membuat DataFrame dari daftar total_penjualan_per_bulan
total_penjualan_per_bulan_2022_df = pd.DataFrame(total_penjualan_per_bulan_2022)
print("Total Penjualan per Bulan untuk Tahun 2022:")
total_penjualan_per_bulan_2022_df.head(12)
```

```python
#Menunjukkan hasil penjualan per bulan dalam bentuk diagram
import matplotlib.pyplot as plt

#Daftar warna yang akan digunakan untuk setiap bulan
warna = ['#7fb3d5', '#afcad5', '#9ed5c8', '#b8a1a1', '#d8a1a1', '#c8b3d5', '#a1c8b8', '#d5c8af', '#d5b3a1', '#c8af9e', '#d5afc8', '#b3d5af']

#Diagram
plt.figure(figsize=(15, 6))
bars = plt.bar(total_penjualan_per_bulan_2022_df['Bulan'], total_penjualan_per_bulan_2022_df['Total Penjualan'], color=warna)
plt.xlabel('Bulan', fontsize=12, fontweight='bold')
plt.ylabel('Penjualan', fontsize=12, fontweight='bold')
plt.title('Total Penjualan per Bulan pada Tahun 2022', fontsize=14, fontweight='bold')
plt.xticks(range(1, 13))
plt.grid(axis='y', linestyle='--', alpha=0.7)

#Menambahkan label jumlah di atas setiap batang
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(int(bar.get_height())), ha='center', va='bottom')

#Penyesuaian sumbu y untuk menarik grafik naik turunnya
plt.ylim(0, max(total_penjualan_per_bulan_2022_df['Total Penjualan'])*1.1)

plt.show()
```

### Peramalan Jumlah Produksi per Bulan pada Tahun 2024

```python
#Data historis penjualan per bulan tahun 2022-2023
penjualan_2022 = [147095, 132549, 147227, 149711, 130815, 145937, 141382, 151132, 141411, 149465, 141160, 156520]
penjualan_2023 = [152221, 135789, 150609, 129289, 152355, 136175, 147938, 149041, 145477, 146357, 141949, 155055]

#Menghitung rata-rata penjualan setiap bulan tahun 2022-2023
rata_rata_penjualan_2022 = sum(penjualan_2022) / len(penjualan_2022)
rata_rata_penjualan_2023 = sum(penjualan_2023) / len(penjualan_2023)

#Lihat hasil rata-rata penjualan
print("Rata-rata penjualan tahun 2022:", rata_rata_penjualan_2022)
print("Rata-rata penjualan tahun 2023:", rata_rata_penjualan_2023)

#Asumsikan tingkat pertumbuhan linier sederhana dari tahun 2023 hingga 2024
tingkat_pertumbuhan = (rata_rata_penjualan_2023 - rata_rata_penjualan_2022) / rata_rata_penjualan_2023
#Lihat hasil tingkat pertumbuhan
print("Tingkat pertumbuhan:", tingkat_pertumbuhan)

#Menerapkan tingkat pertumbuhan untuk memprediksi jumlah produksi pada tahun 2024
peramalan_jumlah_produksi_2024 = [int(sales * (1 + tingkat_pertumbuhan)) for sales in penjualan_2023]
print("Peramalan Jumlah Produksi pe Bulan pada Tahun 2024", peramalan_jumlah_produksi_2024)
```

```python
#Menunjukkan grafik penjualan per bulan
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
months = range(1, 13)
plt.plot(months, peramalan_jumlah_produksi_2024, marker='o', color='blue', label='Peramalan Jumlah Produksi pada Tahun 2024')
plt.xlabel('Bulan', fontsize=12, fontweight='bold')
plt.ylabel('Produksi', fontsize=12, fontweight='bold')
plt.title('Grafik Peramalan Jumlah Produksi per Bulan pada Tahun 2024', fontsize=14, fontweight='bold')
plt.xticks(months)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

### Naive-Bayes Algorithm

```python
#Mengimpor pustaka atau modul bernama pandas ke dalam program Python
import pandas as pd

#Membaca file Excel menggunakan pandas
data = pd.read_excel('Peramalan Jumlah Produksi 2024.xlsx')
data.head(12)
```

```python
#Mengimpor kelas GaussianNB dari pustaka (Gaussian = distribusi normal)
from sklearn.naive_bayes import GaussianNB

#Memisahkan fitur (variabel independen) dan target (variabel dependen)
x = data[['Bulan']]
y = data['Peramalan Jumlah Produksi']

#Membuat model Naive Bayes untuk regresi menggunakan GaussianNB
model = GaussianNB()

#Melatih model
model.fit(x, y)

#Melakukan peramalan
peramalan = model.predict(x)
print("Hasil Peramalan:", peramalan)

#Metrik Evaluasi
#Mengimpor dua fungsi dari pustaka
from sklearn.metrics import mean_squared_error, r2_score

#Data aktual dan data peramalan
data_aktual = [152221, 135789, 150609, 129289, 152355, 136175, 147938, 149041, 145477, 146357, 141949, 155055]
data_peramalan = [152906, 136400, 151287, 129871, 153041, 136788, 148604, 149712, 146132, 147016, 142588, 155753]

#Menghitung MSE
mse = mean_squared_error(data_aktual, data_peramalan)
#Menghitung variabilitas data aktual (variansi)
variasi_data_aktual = plt.np.var(data_aktual)
# Menghitung persentase MSE terhadap variabilitas data aktual
persentase_mse = (mse / variasi_data_aktual)*100
print("Mean Squared Error (MSE):", mse)
print("Variabilitas Data Aktual:", variasi_data_aktual)
print("Persentase MSE terhadap Variabilitas Data Aktual:", persentase_mse)

#Menghitung R2
r2 = r2_score(data_aktual, data_peramalan)
print("R-squared (R2):", r2)
```
